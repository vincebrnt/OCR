## Read Image (ICL Logo)


```python
Image('Imperial_College_Logo.png')
```

```python
read=easyocr.Reader(['en'], gpu=False)
```

## Results

```python
result= read.readtext('Imperial_College_Logo.png')
result
```

## Result  for Multiple Lines (ICL Logo)

```python
top_left=tuple(result[0][0][0])
bottom_right=tuple(result[0][0][2])
text= result[0][1]
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
img=cv2.imread('Imperial_College_Logo.png')
spacer = 20
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.7,(0,255,0),2,cv2.LINE_AA)
    spacer+=20

plt.figure(figsize = (50,5))
plt.imshow(img)
plt.show()
```
